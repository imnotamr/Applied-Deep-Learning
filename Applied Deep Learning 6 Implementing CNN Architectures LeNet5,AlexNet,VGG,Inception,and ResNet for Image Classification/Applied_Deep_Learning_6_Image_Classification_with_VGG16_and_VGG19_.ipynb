# **❏ VGG16**

"VGG16 is a convolutional neural network architecture with 16 layers designed for image classification tasks. It features small 3×3 filters and deep convolutional blocks, followed by fully connected layers. VGG16 is known for its balance of computational efficiency and high accuracy, making it a popular choice for transfer learning and feature extraction in computer vision applications."


    Total Layers: 16 layers (13 convolutional + 3 fully connected layers).
    Convolutional Layers: Divided into 5 blocks, each followed by a max pooling layer.
    Filters: Small 3×3 filters with stride 1 and padding.
    Pooling Layers: Max pooling with 2×2 window and stride 2.
    Parameters: ~138 million.

vgg16.png

```python
from keras.preprocessing.image import load_img, img_to_array
from keras.applications.vgg16 import preprocess_input, decode_predictions
from keras.applications.vgg16 import VGG16
import numpy as np
from IPython.display import display, Image

def predict_image(image_path, description):
    # Load image from file
    image = load_img(image_path, target_size=(224, 224))  # 224x224 pixels for VGG

    # Convert image pixels to a NumPy array
    image_array = img_to_array(image)

    # Reshape data to fit the VGG model input
    image_array = image_array.reshape((1, image_array.shape[0], image_array.shape[1], image_array.shape[2]))

    # Prepare image for the VGG model
    image_array = preprocess_input(image_array)

    # Load the VGG16 model
    model = VGG16()

    # Predict the probability across all output classes
    y_pred = model.predict(image_array)

    # Convert the probabilities to class labels
    label = decode_predictions(y_pred)

    # Retrieve the most likely result which has the highest probability
    label = label[0][0]

    # Print the classification result
    print(f"{description}:")
    print(f"Predicted: {label[1]} ({label[2] * 100:.2f}%)")
    print("-" * 50)

    # Display the image
    display(Image(image_path, width=224, height=224))


images = [
    ('/content/cats.jpg', 'Prediction for Cat Image'),
    ('/content/clothes.jpg', 'Prediction for Clothes Wardrobe Image'),
    ('/content/dog.jpeg', 'Prediction for Dog (golden retriever type) Image'),
    ('/content/park.jpg', 'Prediction for Park Image'),
    ('/content/car.png', 'Prediction for Car Image')
]

# Predict and display for each image
for image_path, description in images:
    predict_image(image_path, description)
```

# **❏ VGG19**

"VGG19 is a deeper convolutional neural network architecture with 19 layers, extending the VGG16 model by adding three additional convolutional layers. Like VGG16, it uses small 3×3 filters and deep convolutional blocks to capture intricate features, achieving high accuracy in image classification tasks. VGG19 is suited for scenarios requiring deeper feature extraction at the cost of slightly higher computational demands."


    Total Layers: 19 layers (16 convolutional + 3 fully connected layers).
    Convolutional Layers: Similar to VGG16 but with 3 extra convolutional layers.
    Filters and Pooling: Same as VGG16 (3×3 filters and 2×2 max pooling).
    Parameters: ~144 million (slightly more than VGG16).
    
vgg19.png
